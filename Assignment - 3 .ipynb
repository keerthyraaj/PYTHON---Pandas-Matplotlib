# ---------------------------------- Fraud Detected Report  ------------------------------------------

# Business Rules:

# 1. In a Trancaction, if Card Type is Signature and  Trancaction Amount lesser than 5000 is termed as Fraudulent Trancaction. 

# 2. In a Trancaction, if Card Type is Silver and  Trancaction Amount greater than 500000 is termed as Fraudulent Trancaction. 

```python
fraud1 = (df['Card Type'] == 'Signature') & (df['Amount'] < 5000)
Fraud_1 =  df[fraud1];Fraud_1
```

```python
fraud2 = (df['Card Type'] == 'Silver') & (df['Amount'] > 500000)
Fraud_2 =  df[fraud2];Fraud_2
```

```python
frames4 = [Fraud_1, Fraud_2]
Fraud_Detected_Report = pd.concat(frames4)
Fraud_Detected_Report
```
